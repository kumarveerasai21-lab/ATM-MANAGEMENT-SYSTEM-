```python
from abc import ABC, abstractmethod

# Abstract class
class BankAccount(ABC):
    @abstractmethod
    def transaction(self):
        pass


# Base class (Encapsulation)
class Account(BankAccount):
    def __init__(self, name, acc_no, balance):
        self.name = name              # public
        self._acc_no = acc_no         # protected
        self.__balance = balance      # private

    def show_name(self):
        print("Account Holder Name:", self.name)

    def _show_account_number(self):
        print("Account Number:", self._acc_no)

    def __show_balance(self):
        print("Account Balance:", self.__balance)

    def show_balance(self):
        self.__show_balance()

    def transaction(self):
        print("Basic bank transaction")


# Child class (Inheritance + Polymorphism)
class ATM(Account):
    bank_name = "ABC Bank"

    # Class method
    @classmethod
    def show_bank_name(cls):
        print("WELCOME TO", cls.bank_name, "ATM")

    # Overridden method
    def transaction(self):
        print("--------------------------------")
        ATM.show_bank_name()
        print("--------------------------------")
        print("Account Holder:", self.name)
        print("Account Number:", self._acc_no)

    # Static method
    @staticmethod
    def is_valid_amount(amount):
        return amount > 0

    # Check balance
    def check_balance(self):
        print("Current Balance:", self._Account__balance)

    # Deposit money
    def deposit(self, amount):
        if ATM.is_valid_amount(amount):
            self._Account__balance += amount
            print("Deposit successful")
            print("Updated Balance:", self._Account__balance)
        else:
            print("Invalid amount")

    # Withdraw money
    def withdraw(self, amount):
        if ATM.is_valid_amount(amount):
            if amount <= self._Account__balance:
                self._Account__balance -= amount
                print("Withdrawal successful")
                print("Updated Balance:", self._Account__balance)
            else:
                print("Insufficient balance")
        else:
            print("Invalid amount")

    # Account details
    def account_details(self):
        self.show_name()
        self._show_account_number()
        self.show_balance()


# Object creation
atm = ATM("Ganesh Koduri", 6303135752, 10000)

atm.transaction()

# Menu-driven program
while True:
    print("\nATM MENU")
    print("1. Check Balance")
    print("2. Deposit Money")
    print("3. Withdraw Money")
    print("4. Account Details")
    print("5. Exit")
    print("----------------")

    choice = input("Enter your choice: ")

    if choice == "1":
        atm.check_balance()

    elif choice == "2":
        amt = float(input("Enter deposit amount: "))
        atm.deposit(amt)

    elif choice == "3":
        amt = float(input("Enter withdrawal amount: "))
        atm.withdraw(amt)

    elif choice == "4":
        atm.account_details()

    elif choice == "5":
        print("Thank you for using", ATM.bank_name)
        break

    else:
        print("Invalid choice")
```
